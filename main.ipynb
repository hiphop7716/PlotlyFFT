```python
import csv
import pandas as pd
import numpy as np

import plotly.offline as py
import plotly.graph_objs as go
py.init_notebook_mode(connected=True)

df_1 = pd.read_csv(r'PlotlyFFT/cp/fft_1.csv')
df_2 = pd.read_csv(r'PlotlyFFT/cp/fft_2.csv')
df_3 = pd.read_csv(r'PlotlyFFT/cp/fft_3.csv')
df_4 = pd.read_csv(r'PlotlyFFT/cp/fft_4.csv')

# df = pd.read_csv(r'cp\fft_1.csv')
# print(df.head(3))

x1 = df_1.loc[:, 'frequency']
y1 = df_1.loc[:, 'x']
x2 = df_2.loc[:, 'frequency']
y2 = df_2.loc[:, 'x']
x3 = df_3.loc[:, 'frequency']
y3 = df_3.loc[:, 'x']
x4 = df_4.loc[:, 'frequency']
y4 = df_4.loc[:, 'x']

trace1 = go.Scatter(
    x = x1,
    y = y1
)
trace2 = go.Scatter(
    x = x2,
    y = y2
)
trace3 = go.Scatter(
    x = x3,
    y = y3
)
trace4 = go.Scatter(
    x = x4,
    y = y4
)


data = [trace1, trace2, trace3, trace4]

layout = go.Layout(
    xaxis = go.layout.XAxis(
        tickmode = 'linear',
        tick0 = 0,
        dtick = 50
    )
)

fig = go.Figure(
    data = data,
    layout = layout
)

py.iplot(fig, filename='tickformatting-tickmode-linear')


x1 = df_1.loc[:, 'frequency']
y1 = df_1.loc[:, 'y']
x2 = df_2.loc[:, 'frequency']
y2 = df_2.loc[:, 'y']
x3 = df_3.loc[:, 'frequency']
y3 = df_3.loc[:, 'y']
x4 = df_4.loc[:, 'frequency']
y4 = df_4.loc[:, 'y']

trace1 = go.Scatter(
    x = x1,
    y = y1
)
trace2 = go.Scatter(
    x = x2,
    y = y2
)
trace3 = go.Scatter(
    x = x3,
    y = y3
)
trace4 = go.Scatter(
    x = x4,
    y = y4
)

data = [trace1, trace2, trace3, trace4]

layout = go.Layout(
    xaxis = go.layout.XAxis(
        tickmode = 'linear',
        tick0 = 0,
        dtick = 50
    )
)

fig = go.Figure(
    data = data,
    layout = layout
)

py.iplot(fig, filename='tickformatting-tickmode-linear')


x1 = df_1.loc[:, 'frequency']
y1 = df_1.loc[:, 'z']
x2 = df_2.loc[:, 'frequency']
y2 = df_2.loc[:, 'z']
x3 = df_3.loc[:, 'frequency']
y3 = df_3.loc[:, 'z']
x4 = df_4.loc[:, 'frequency']
y4 = df_4.loc[:, 'z']

trace1 = go.Scatter(
    x = x1,
    y = y1
)
trace2 = go.Scatter(
    x = x2,
    y = y2
)
trace3 = go.Scatter(
    x = x3,
    y = y3
)
trace4 = go.Scatter(
    x = x4,
    y = y4
)

data = [trace1, trace2, trace3, trace4]

layout = go.Layout(
    xaxis = go.layout.XAxis(
        tickmode = 'linear',
        tick0 = 0,
        dtick = 50
    )
)

fig = go.Figure(
    data = data,
    layout = layout
)

py.iplot(fig, filename='tickformatting-tickmode-linear')
```
